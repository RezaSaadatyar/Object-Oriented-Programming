```python
class human:
    def __init__(self, name, lname, age):
        self.name = name 
        self.lname = lname
        self.age = age
    def fun(self):
        return f"{self.name}"

a = human(name="ali", lname="kf",age=20)
print(a.lname)
print(a.fun())
```

```python
class student:
    users = []                     # Class variable to store list of user names
    users_course = {}              # Class variable to store dictionary mapping user names to their courses
    
    def __init__(self, name, email, password, recive):
        self.name = name           # Instance variable to store the student's name
        self.email = email         # Instance variable to store the student's email
        self.password = password   # Instance variable to store the student's password
        self.recive = recive       # Instance variable to store whether the student wants to receive updates
        # print(f"{self.name} wellcome!")  # Greeting print statement, commented out
        student.users.append(self.name)  # Adds the new student's name to the class variable list `users`
    
    def login(self, name):
        if name in student.users:  # Checks if the given name is in the `users` list
            print(f"{name} have an acount.")  # Prints that the user has an account
        else:
            print(f"{name} must be register.")  # Prints that the user must register
            
    def buy(self, coursename):
        student.users_course[self.name] = coursename  # Assigns a course to the user in the `users_course` dictionary
        
output = student("aa","a@.com","2547", True)  # Creates an instance of `student` with the name "aa"
output.buy("matlab")  # Calls the `buy` method to assign the course "matlab" to user "aa"
output = student("rr","r@.com","247", True)  # Creates another instance of `student` with the name "rr"
output.buy("python")  # Calls the `buy` method to assign the course "python" to user "rr"
print(output.users)  # Prints the list of users; should show ['aa', 'rr']
print(output.users_course)  # Prints the dictionary of users and their courses; should show {'aa': 'matlab', 'rr': 'python'}
output.login("f")  # Attempts to log in a user named "f", should print "f must be register."
output = student("f", "f@.com", "5478", True)  # Creates a new instance of `student` with the name "f"
output.login("f")  # Now attempts to log in "f" again, should print "f have an acount."
print(output.users)  # Prints the list of users; should now include "f", showing ['aa', 'rr', 'f']
```
