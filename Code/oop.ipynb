    - `__init__` is a special method in Python known as a constructor. It is automatically called when an instance of the class is created.<br/>

```python
class human:  # Define a class named `human`
    def __init__(self, name, lname, age):
        self.name = name  # Initialize `name` attribute with the provided first name
        self.lname = lname  # Initialize `lname` attribute with the provided last name
        self.age = age  # Initialize `age` attribute with the provided age value

    def fun(self):
        return f"{self.name}"  # Method that returns the first name of the human instance

# Create an object of the class `human` with specified attributes (name, lname, age)
a = human(name="ali", lname="kf", age=20)
# Access attributes
print(f"{a.age = }")  # Print the age of the object 'a' (age = 20)
print(f"{a.name = }")  # Print the name of the object 'a' (name = 'ali')
print(f"{a.lname = }")  # Print the last name of the object 'a' (lname = 'kf')

print(f"{a.fun() = }")  # Calls the 'fun' method of instance 'a' and prints the return value ("ali")
```

```python
class student:
    users = []                   # Class variable to store list of user names
    users_course = {}            # Class variable to store dictionary mapping user names to their courses
    
    def __init__(self, name, email, password, receive):
        self.name = name         # Instance variable to store the student's name
        self.email = email       # Instance variable to store the student's email
        self.password = password # Instance variable to store the student's password
        self.receive = receive   # Instance variable to store whether the student wants to receive updates
        student.users.append(self.name)  # Adds the new student's name to the class variable list `users`
        print(f"{self.name} welcome!")  # Greeting the user upon creation
    
    def login(self, name):
        if name in student.users:  # Checks if the given name exists in the `users` list
            print(f"{name} has an account.")  # Indicates the user has an account
        else:
            print(f"{name} must register.")  # Indicates the user must register
            
    def buy(self, coursename): # Method to assign a course to the student
        student.users_course[self.name] = coursename  # Maps the student's name to the purchased course
        print(f"{self.name} has purchased the course: {coursename}")

# Create an object of the class
student_aa = student("aa", "a@.com", "2547", True)  # Creates an instance of `student` named "aa"
student_aa.buy("matlab")  # The student "aa" purchases the "matlab" course
student_rr = student("rr", "r@.com", "247", True)  # Creates another instance of `student` named "rr"
student_rr.buy("python")  # The student "rr" purchases the "python" course
print(student.users)  # Prints the list of user names, expected output: ['aa', 'rr']
print(student.users_course)  # Prints the dictionary of users and their assigned courses
student_rr.login("f")  # Attempts to log in a user named "f", expected print: "f must register."
student_f = student("f", "f@.com", "5478", True)  # Creates a new instance of `student` named "f"
student_f.login("f")  # Now attempts to log in "f" again, expected print: "f has an account."
print(student.users)  # Prints the updated list of user names, now including "f"
```

```python
class student:
    users = []                   # Class variable that keeps a list of user names
    users_course = {}            # Class variable that maps user names to their purchased courses
    
    def __init__(self, name, email, password, receive):
        self.name = name         # Instance variable for storing the student's name
        self.email = email       # Instance variable for storing the student's email
        self.password = password # Instance variable for storing the student's password
        self.receive = receive   # Instance variable for storing whether the student opts to receive notifications
        student.users.append(self.name)  # Adds the student's name to the class list `users`
        print(f"{self.name} is added.")  # Prints a message confirming the addition of the student
    
    # def __del__(self): # Removes user from the `users` list upon object deletion (currently commented out)
    #     student.users.remove(self.name)  
    
    def remove_user(self):  # Method to explicitly remove a user
        student.users.remove(self.name)  # Removes the student from the `users` list
        print(f"{self.name} is removed.")  # Prints a confirmation that the user has been removed
        if self.name in student.users_course:
            del student.users_course[self.name]  # Removes any course associated with the student from `users_course`
            print(f"course {self.name} is removed.")  # Prints a confirmation that the course registration is removed

    def login(self, name):  # Method for logging in a user
        if name in student.users:  # Checks if the name exists in the `users` list
            print(f"{name} has an account.")  # Confirmation message that the user has an account
        else:
            print(f"{name} must register.")  # Message indicating the user needs to register
            
    def buy(self, coursename):  # Method to assign a course to a student
        student.users_course[self.name] = coursename  # Maps the course name to the student in `users_course`
        print(f"{self.name} bought {coursename}.")  # Prints a confirmation that the course has been purchased


# Create an object of the class
student_jame = student("Jame", "a@.com", "2547", True)  # Creates an instance `student_jame`
student_jame.buy("matlab")  # `student_jame` buys the course "matlab"
student_sara = student("Sara", "r@.com", "247", True)  # Creates another instance `student_sara`
student_sara.buy("python")  # `student_sara` buys the course "python"
student_f = student("f", "f@.com", "5478", True)  # Creates another instance `student_f`
student_f.login("f")  # Attempts to log in the student "f"
print(f"{student.users = }")  # Prints the current list of all users
print(f"{student.users_course = }")  # Prints the current mapping of users to their courses

student_jame.remove_user()  # Calls the method to remove `student_jame` from users
print(f"{student.users = }")  # Prints the updated list of users
print(f"{student.users_course = }")  # Prints the updated mapping of users to their courses
```

- Representation Method `__repr__`
  - `__repr__` is a special method used to define the string representation of an object. It returns a string that describes the object.

```python
class human:
    def __init__(self, name, lname, age):
        self.name = name       # Instance variable for first name
        self.lname = lname     # Instance variable for last name
        self.age = age         # Instance variable for age
    
    def __repr__(self):
        # Representation method that returns a string format of the human instance
        return f"{self.name} {self.lname}: {self.age}"

# Create an object of the class of human with specified attributes
output = human(name="Sara", lname="kf", age=20)
print(output)  # Prints the representation of the instance 'output', which is "Sara kf: 20"
```

**Inheritance**
- Constructor Method `__init__`
- Method
  - sleep
- Student Class (Inherits from person)
  - The student class inherits from person, meaning it has access to the methods and properties of person, while also adding its own attributes and methods.
- super().`__init__`
  - The super() method is generally preferred for better maintainability and to support multiple inheritance.
- Method
  - score

```python
class person:
    def __init__(self, name, lname, age):
        self.name = name       # Store the person's first name
        self.lname = lname     # Store the person's last name
        self.age = age         # Store the person's age
    
    def sleep(self):
        # Return a string indicating that the person is sleeping
        return f"{self.name} is sleeping ..."

class student(person):
    def __init__(self, name, lname, age, student_id):
        # Initialize the parent class (person) with the provided name, last name, and age
        person.__init__(self, name, lname, age)  # or super().__init__(name, lname, age)
        self.student_id = student_id  # Store the student's ID
    
    def score(self):
        # Return a string indicating that the student received a score
        return f"{self.name} received a score."

# Create an object of the class 
output = student("Jame", "hfgddh", 25, 12547) 
print(f"output.student_id = {output.student_id}") # Print the student ID of 'output'
print(f"output.sleep() = {output.sleep()}") # Print the result of the 'sleep' method called on 'output'
print(f"output.score() = {output.score()}") # Print the result of the 'score' method called on 'output'
```

- *arg
  - It allows a function to accept any number of positional arguments. The arguments passed with *args are stored in a tuple.
- **kwarg
  - It allows a function to accept any number of keyword arguments (arguments passed with a key-value pair). The arguments passed with **kwargs are stored in a dictionary.

```python
def example(*args):
    for arg in args:
        print(arg)

example(1, 2, 3, 4, 5)
```

```python
def example(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

example(a=1, b=2, c=3)
```

```python
class person:
    def __init__(self, name, lname, age):
        self.name = name   # Store the person's first name
        self.lname = lname # Store the person's last name
        self.age = age     # Store the person's age
        
    def sleep(self):
        # Returns a string indicating the person is sleeping
        return f"{self.name} is sleeping ..."

class student(person):
    def __init__(self, student_id, *args, **kwargs):
        # Initialize the parent class with additional arguments
        super().__init__(*args, **kwargs)  # Using super() is a more modern and clear approach
        self.student_id = student_id  # Store the student's ID
    
    def score(self):
        # Returns a string indicating that the student received a score
        return f"{self.name} received a score."

# Create an instance of 'student' with named arguments
output = student(name="Jame", lname="hfgddh", age=25, student_id=12547)
print(f"output.student_id = {output.student_id}")  # Prints the student's ID
print(f"output.sleep() = {output.sleep()}")        # Prints the sleep message
print(f"output.score() = {output.score()}")        # Prints the score message
```

**Polymorphism or Overloading**<br/>
It allows a single interface (method or property) to be used for different underlying data types. In other words, it lets you call the same method on different objects, and each object can respond in its own way.

```python
class parent:
    def fun(self):
        # Prints a message indicating this method belongs to the parent class
        print("Parent class")

class child(parent):
    def fun(self):
        # Overrides the parent class's 'fun' method, printing a different message
        print("Child class")

# Create an instance of the parent class
p = parent()
# Create an instance of the child class
c = child()

# Call the 'fun' method on an instance of the parent
p.fun()  # Output: Parent class
# Call the overridden 'fun' method on an instance of the child
c.fun()  # Output: Child class
```

```python
class Circle:
    def draw(self):
        # This method prints a message specific to drawing a circle
        print("Drawing a circle.")

class Square:
    def draw(self):
        # This method prints a message specific to drawing a square
        print("Drawing a square.")
        
# Create an instance of the Circle class
c = Circle()
# Create an instance of the Square class
s = Square()

# Invoke the draw method on the Circle instance
c.draw()  # Expected output: "Drawing a circle."
# Invoke the draw method on the Square instance
s.draw()  # Expected output: "Drawing a square."
```

**Decorator**<br/>
A decorator is a function that takes another function as an argument and returns a new function that usually extends or modifies the behavior of the original function.<br/>
Apply the Decorator: Using the `@decorator_name` syntax above the function definition to apply the decorator.

```python
def func1():
    print("Function 1")
    
    def func2():
        print("Function 2")
    
    def func3():
        print("Function 3")
    
    return func2()

f = func1()
f
```

```python
# Define a decorator
def greet_decorator(function):
    def wrapper():
        print("Hello there!")
        function()
    return wrapper

def func():
    print("Fun")

f = greet_decorator(func)
f()
```

```python
# Define a decorator
def greet_decorator(function):
    def wrapper():
        print("Hello there!")
        function()
    return wrapper

@greet_decorator
def func():
    print("Fun")
func()
```

```python
# Define a decorator
def my_decorator(func):
    def wrapper(*args, **kwargs):
        print("Something is happening before the function is called.")
        result = func(*args, **kwargs)
        print("Something is happening after the function is called.")
        return result
    return wrapper

# Use the decorator
@my_decorator
def say_hello(name):
    print(f"Hello, {name}!")

# Call the decorated function
say_hello("Alice")
```

**@dataclass**<br/>
The @dataclass decorator is a feature from the dataclasses module that simplifies the creation of classes used primarily to store data. It automatically generates special methods for classes, such as `__init__`, `__repr__`, `__eq__`, and others, reducing boilerplate code.<br/>

The @dataclass decorator in Python automatically provides several features:
- `__init__` Method: Automatically generates an initializer based on class attributes.
- `__repr__` Method: Generates a string representation of the instance that includes its attribute values.
- `__eq__` Method: Implements comparison for equality between instances based on their attribute values.
- `__hash__` Method: Provides a hash function for instances, useful if instances are used as keys in dictionaries or sets.
- `__post_init__` Method: Allows you to define custom initialization logic after the default __init__ method is called.


```python
# -------------------------------------- Without @dataclass ----------------------------------------------
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def __repr__(self):
        return f"Person(name={self.name!r}, age={self.age!r})"
    
    def __eq__(self, other):
        if isinstance(other, Person):
            return self.name == other.name and self.age == other.age
        return False

p = Person("Sara", 32)
print(f"{p.name = }, {p.age = }")
```

```python
# ---------------------------------------- With @dataclass -----------------------------------------------
from dataclasses import dataclass

@dataclass
class Person:
    name: str
    age: int
    
p = Person("Sara", 32)
print(f"{p.name = }, {p.age = }")
```
